```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error,mean_absolute_error
```

```python
df = pd.read_csv('student_marks.csv')
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.study_hours.mean()
```

```python
df2 =df.fillna(df.study_hours.mean())
df2
```

```python
X = df2.drop('student_marks', axis=1)
X
```

```python
y= df2.drop('study_hours', axis=1)
y
```

```python
#mean_x = X.mean()
#mean_y = y.mean()
#mean_x,mean_y
```

```python
plt.scatter(X, y)
plt.xlabel("Students Study Hours")
plt.ylabel("Students marks")
plt.title("Scatter Plot of Students Study Hours vs Students marks")
```

```python
X_train,X_test,y_train,y_test= train_test_split(X,y,test_size =0.2,random_state=6)
```

```python
X_train,y_train
```

```python
X_test
```

```python
y_test
```

```python
reg = LinearRegression()
```

```python
reg.fit(X_train,y_train)
```

```python
m=reg.coef_
m
```

```python
c=reg.intercept_
c
```

```python
m*40+c
```

```python
reg.predict([[40]])
```

```python
y_pred= reg.predict(X_test)
y_pred
```

```python
final_df=pd.DataFrame(np.c_[X_test, y_test, y_pred], columns = ["study_hours", "student_marks_original","student_marks_predicted"])
final_df
```

```python
reg.score(X_test,y_test)
```
